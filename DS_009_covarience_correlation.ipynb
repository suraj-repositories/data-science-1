```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("dataset/titanic.csv")
```

```python
dataset.head(3)
```

```python
dataset.info()
```

```python
dataset.select_dtypes(["int64", "float64"])
```

```python
data_corr = dataset.select_dtypes(["int64", "float64"]).corr()
data_corr
```

```python
data_cov = dataset.select_dtypes(["int64", "float64"]).cov()
data_cov
```

```python
sns.heatmap(data_corr, annot=True)
plt.show()
```

```python
sns.heatmap(data_cov, annot=True)
plt.show()
```
