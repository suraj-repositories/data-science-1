## z score

- formule

  z = (x - mean) / standard_deviation

  

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"./dataset/loan.csv")
dataset.head(3)
```

```python
dataset.describe()
```

```python
sns.boxplot(x="CoapplicantIncome", data = dataset)
plt.show()
```

```python
dataset.isnull().sum()
```

```python
sns.histplot(dataset["CoapplicantIncome"], kde=True)
plt.show()
```

## Outlier Removal : simple method

```python
min_range = dataset["CoapplicantIncome"].mean() - (3*dataset["CoapplicantIncome"].std())
max_range = dataset["CoapplicantIncome"].mean() + (3*dataset["CoapplicantIncome"].std())
min_range, max_range
```

```python
new_dataset = dataset[dataset["CoapplicantIncome"]<=max_range]
```

```python
sns.boxplot(x="CoapplicantIncome", data = new_dataset)
plt.show()
```

## Outlier Removal : Z-score method

```python
z_score = (dataset["CoapplicantIncome"] - dataset["CoapplicantIncome"].mean()) / dataset["CoapplicantIncome"].std()
```

```python
dataset["z_score"] = z_score
```

```python
new_set = dataset[dataset["z_score"] < 3]
new_set
```

```python
sns.boxplot(x="CoapplicantIncome", data = new_set)
plt.show()
```
