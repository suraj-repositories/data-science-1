```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sec_a = np.array([75, 65, 73, 68, 72, 76])
sec_b = np.array([90, 47, 43, 96, 93, 51])
no = np.array([1, 2, 3, 4, 5, 6])
```

```python
mean = np.mean(sec_b)
```

```python
mad_a = np.sum(np.abs(sec_a - mean)) / len(sec_a)
mad_b = np.sum(np.abs(sec_b - mean)) / len(sec_b)
```

```python
mad_a, mad_b
```

```python
plt.figure(figsize=(10, 3))
plt.scatter(sec_a, no ,label="sec A")
plt.scatter(sec_b, no, label="sec B")
plt.axvline(mean, color="red", linestyle="--", linewidth=2, label="Mean")

plt.legend()
plt.show()
```
