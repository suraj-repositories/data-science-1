```python
# Mode = highest occurence value in dataset is called mode
# mode frequency = how many times the value come in dataset
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("dataset/products-100.csv")
```

```python
dataset.head(3)
```

```python
dataset.fillna({'Stock': dataset['Stock'].mean()}, inplace=True)
```

```python
mo = dataset["Stock"].mode()[0]
```

```python
dataset["Stock"].value_counts()
```

```python
md = np.median(dataset['Price'])
```

```python
mn = np.mean(dataset['Price'])
```

```python
sns.histplot(x="Price", data=dataset, bins=[i for i in range(0, 1000, 50)])

plt.axvline(mn, color="red", linestyle="--", linewidth=2, label="Mean")
plt.axvline(md, color="blue", linestyle="--", linewidth=2, label="Median")
plt.axvline(mo, color="green", linestyle="--", linewidth=2, label="Mode")

plt.legend()
plt.show()
```
