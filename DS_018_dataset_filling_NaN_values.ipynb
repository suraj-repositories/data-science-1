### Data filling 

- for categorical data (string data) we use the mode to fill the values
- for numerical data we use the (mean or median) to fill the values

```python


## filling mode values to object datatype
# 1. find mode
mode = dataset["Gender"].mode()[0]
mode

# 2. fill in a column and replace the column 
dataset["Gender"] = dataset["Gender"].fillna(mode)
dataset.info()
```

```python
## fill all the dataset : of column object type with there mode values

# 1. get the columns of type object
dataset.select_dtypes(include="object").isnull().sum()
```

```python
for i in dataset.select_dtypes(include="object").columns:
     dataset[i] = dataset[i].fillna(dataset[i].mode()[0])
```
