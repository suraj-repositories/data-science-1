## Duplicates : 

when the entire data of a row repeat on another row : called duplicate

```python
import pandas as pd 
```

```python
data = {"name": ["a","b","c","d","a","c"], "eng" : [8,7,5,8,8,5], "hindi": [2,3,4,5,2,6]}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.drop_duplicates()
```
