```python
import pandas as pd
```

```python
df = pd.DataFrame({"size": ["s", "m", "l", "xl", "s", "m", "l", "l", "xl", "m", "s", "xxl", "l"]})
df.head(3)
```

```python
ord_data  =[["s", "m", "l", "xl", "xxl"]]
```

## using scikit LIb

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
oe.fit(df[["size"]])
```

```python
df["size_en"] = oe.transform(df[["size"]])
df
```

## Using Map

```python
ord_data_1 = {"s": 0,"m": 1,"l": 2,"xl": 3,"xxl":4}
```

```python
df["size_en_map"] = df["size"].map(ord_data_1)
```

```python
df
```

## Ordinal Encoding on Large Dataset - 

```python
dataset = pd.read_csv(r"./dataset/loan.csv")
dataset.head(3)
```

```python
dataset["Property_Area"].unique()
```

```python
dataset["Property_Area"] = dataset["Property_Area"].fillna(dataset["Property_Area"].mode()[0])
```

```python
en_data_ord = [['Urban', 'Rural', 'Semiurban']]
```

```python
oen = OrdinalEncoder(categories=en_data_ord)
oen
```

```python
## The fit transform function :: applies fit() then transform() : to it will change the categorical data to numerical data form analysic
dataset["Property_Area"] = oen.fit_transform(dataset[["Property_Area"]])
```
