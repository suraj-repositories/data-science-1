## Outlier : 

A value in dataset which is different from other values : 
for example extreamly high or low then other vales called outlier

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"./dataset/loan.csv")
dataset.head(3)
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
sns.boxplot(x="ApplicantIncome", data=dataset)
plt.show()


## we can see the data outsize of boxplot regin : is called outlier
```

```python
sns.histplot(dataset["ApplicantIncome"], kde=True)
plt.show()
```
