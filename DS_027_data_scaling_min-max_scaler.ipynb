```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
dataset = pd.read_csv(r"./dataset/loan.csv")
dataset.head(3)
```

```python
dataset.describe()
```

```python
ms = MinMaxScaler()
ms.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_min"] = ms.transform(dataset[["CoapplicantIncome"]])
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.title("Before")
sns.histplot(dataset["CoapplicantIncome"], kde=True)

plt.subplot(1,2,2)
plt.title("After")
sns.histplot(dataset["CoapplicantIncome_min"], kde=True)

plt.show()
```
