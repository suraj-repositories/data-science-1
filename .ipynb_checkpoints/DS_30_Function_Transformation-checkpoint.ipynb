```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.preprocessing import FunctionTransformer
import numpy as np
```

```python
dataset = pd.read_csv(r"./dataset/loan.csv")
dataset.head(3)
```

```python
dataset.info()
```

```python
## remove outliers
q1 = dataset["CoapplicantIncome"].quantile(0.25)
q3 = dataset["CoapplicantIncome"].quantile(0.75)

IQR = q3 - q1

min_range = q1 - (1.5 * IQR)
max_range = q3 + (1.5 * IQR)

new_dataset = dataset[dataset["CoapplicantIncome"] <= max_range]
```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.title("Before")
sns.histplot(dataset["CoapplicantIncome"], kde=True)

plt.subplot(1,2,2)
plt.title("After")
sns.histplot(new_dataset["CoapplicantIncome"], kde=True)

plt.show()
```

```python
## starting function transformation
# you can use any function which convert your curve to "Normal distribution curve" - beause modals work well on normal distribution data

ft = FunctionTransformer(func=np.log1p)

# or
# ft = FunctionTransformer(func= lambda x: x**2)

```

```python
ft.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_tf"] = ft.transform(dataset[["CoapplicantIncome"]])
```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.title("Before")
sns.histplot(dataset["CoapplicantIncome"], kde=True)

plt.subplot(1,2,2)
plt.title("After")
sns.histplot(dataset["CoapplicantIncome_tf"], kde=True)

plt.show()
```
