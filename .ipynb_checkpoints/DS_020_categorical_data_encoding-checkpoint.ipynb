```python
import pandas as pd
```

```python
dataset = pd.read_csv(r"./dataset/loan.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
# filling null values first
dataset["Gender"] = dataset["Gender"].fillna(dataset["Gender"].mode()[0])
dataset["Married"] = dataset["Married"].fillna(dataset["Married"].mode()[0])
```

#### Way 1 : Get dummies function

```python
en_data = dataset[["Gender", "Married"]]
```

```python
pd.get_dummies(en_data)
```

```python
pd.get_dummies(en_data).info()
```

#### Way 2 : One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
# ohe = OneHotEncoder(drop="first") ## to reduce the number of columns generated after incoding
ar = ohe.fit_transform(en_data).toarray()

# Sparse matrix - sparse matrix where most of the values are filled with either 0 or 1
```

```python
pd.DataFrame(ar, columns=["Gender_Female", "Gender_Male", "Married_No", "Married_Yes"])
```
