```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"./dataset/loan.csv")
dataset.head(3)
```

```python
dataset.info()
```

```python
dataset.select_dtypes(include = "float64").columns
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(strategy="mean")
ar = si.fit_transform(dataset[['CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term',
       'Credit_History']])

## fit transform fill the given column with the simple imputer strategy
```

```python
# we can convert our : changed data to dataset again by using dataframe
new_dataset = pd.DataFrame(ar, columns=dataset.select_dtypes(include = "float64").columns)
```

```python
# now it doesn't have null values
new_dataset.isnull().sum()
```
