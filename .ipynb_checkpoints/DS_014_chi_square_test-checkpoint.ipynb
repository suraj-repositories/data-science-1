## Chi Square Test

### Question: 
A fair dice is rolled 120 times and the followign result are obtainged.
- Face 1: 22 times -> 20
- Face 2: 17 times -> 20
- Face 3: 20 times -> 20
- Face 4: 26 times -> 20
- Face 5: 22 times -> 20
- Face 6: 13 times -> 20

Test at a 5% level of significance whether the dice is fare.




### Answer

Step 1: Write Hypotheses

Null Hypothesis (H₀): The die is fair.

Alternative Hypothesis (H₁): The die is not fair.

If the die is fair, each face should occur equally often → Expected frequency = 
120/6=20

```python
import numpy as np
from scipy.stats import chi2
```

```python
ob = np.array([22,17,20,26,22,13]) # observation values
ex = np.array([20,20,20,20,20,20]) # expected values
```

```python
critical_value = chi2.ppf(1 - 0.05, df=5)
critical_value
```

```python
chi_square_test = np.sum(np.square(ob-ex)/ex)
chi_square_test
```

```python
if chi_square_test < critical_value :
    print("Ho is fair")
else:
    print("h1 is right , die is unfare")
```
