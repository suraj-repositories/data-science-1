`Question:` A company wants to test whether there is a difference in productivity between two teams. They randomly select 20 employees from each team and add record their productivity scores. The mean productivity score for team A is 80 with a standard deviation of 5, while the mean productvity score for team B is 75 with the standard deviation of 6. Test at a 5% level of significance whether there is a difference produtivity between the tow teams.

let <br> 
h0 = both are same<br>
h1 = both are different

```python
import numpy as np
import scipy.stats as st
```

```python
degree_of_freedom = 20 + 20 - 2
```

```python
t_table = st.t.ppf(1 - 0.025, degree_of_freedom)
t_table
```

```python
t_cal = (80-75)/np.sqrt((25/20) + (26/20))
t_cal
```

```python
if(t_cal > t_table or t_cal < t_table):
    print("h1 is right")
else:
    print("h0 is right")
```
