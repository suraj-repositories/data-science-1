### Question : 

A study was conducted to investigate whether there is a relationship between gender and the preferred genre of music. A sample of 235 people was selected and the data collected to shown below. Test at a 5% level of significance whether there is a significant association between gender and music preference.


|        | Pop | Hip Hop | Classical | Rock |
|--------|-----|----------|------------|------|
| Male   | 40  | 45       | 25         | 10   |
| Female | 35  | 30       | 20         | 30   |


### Answer : 

```python
from scipy.stats import chi2
import numpy as np
```

```python
alpha = 0.05
# degree_of_freedom = (row - 1) * (col - 1)
df = (2 - 1) * (4 - 1)
df
```

```python
cv = chi2.ppf(1 - 0.05, df=df)
cv
```

```python
row1 = np.array([40,45,25,10])
row2 = np.array([35,30,20,30])
```

```python
sum_r1 = np.sum(row1)
sum_r2 = np.sum(row2)
sum_row = np.array([sum_r1, sum_r2])
sum_row
```

```python
sum_col = row1 + row2
sum_col
```

```python
exp = []
for i in sum_row:
    for j in sum_col:
        value = (i * j) / 235
        exp.append(value)


exp # expected value
```

```python
obv = np.array([40,45,25,10,35,30,20,30])
obv # observation value
```

```python
chi_calc = np.sum(np.square(obv - exp) / exp)
chi_calc 
```

```python
if chi_calc < cv :
    print("Ho is fair")
else:
    print("h1 is right , there are significant associatin between genders")
```
