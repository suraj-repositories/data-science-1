**Question :**
A company wants to test whether a new training program improves the typing speed of its employees. The typing speed of 20 employees was recorded before and after the training program. The data is given below. Test at a 5% level of significance whether the trainging has an effect on the typing speed of the employees.

- Before = 50,60,45,65,55,70,40,75,80,65,70,60,50,55,45,75,60,50,65,70
- After = 60,70,55,75,65,80,50,85,90,70,75,65,55,60,50,80,65,55,70,75

**Answer :** 
let  
- null hypothysis : h0 = no difference
- h1 = after > before

```python
import scipy.stats as st
import numpy as np
```

```python
t_table = st.t.ppf(1.-0.025,19)
t_table
```

```python
before = np.array([50,60,45,65,55,70,40,75,80,65,70,60,50,55,45,75,60,50,65,70])
after = np.array([60,70,55,75,65,80,50,85,90,70,75,65,55,60,50,80,65,55,70,75])
```

```python
std_a = np.std(after)
std_b = np.std(before)
mean_a = np.mean(after)
mean_b = np.mean(before)
```

```python
t_cal = (mean_a - mean_b) / (np.sqrt(((std_a*std_a)/len(after))+((std_b*std_b)/len(before))))
t_cal
```

```python
if t_cal > t_table: 
    print("h1 is right : significant difference")
else: 
    print("h0 is right : no significant difference")
```
