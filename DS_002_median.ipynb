```python
# Median = Center position of the data
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("dataset/products-100.csv")
```

```python
dataset.head(3)
```

```python
dataset.fillna({'Price': dataset['Price'].mean()}, inplace=True)
```

```python
dataset['Price'].median()
```

```python
md = np.median(dataset['Price'])
```

```python
mn = np.mean(dataset['Price'])
```

```python
sns.histplot(x="Price", data=dataset, bins=[i for i in range(0, 1000, 50)])
plt.axvline(mn, color="red", linestyle="--", linewidth=2) 
plt.axvline(md, color="blue", linestyle="--", linewidth=2) 
plt.show()
```
